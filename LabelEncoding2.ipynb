```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
iris_data['Species_encoded'] = label_encoder.fit_transform(iris_data['Species'])
iris_data.head()
```

```python
iris_data['Species_encoded'].value_counts()
```

```python
iris_setosa = iris_data[iris_data.Species_encoded == 0]
iris_virginica = iris_data[iris_data.Species_encoded == 1]
iris_versi_color = iris_data[iris_data.Species_encoded == 2]
```

```python
print(iris_setosa.shape)
print(iris_virginica.shape)
print(iris_versi_color.shape)
```

```python
setosa_sample = iris_setosa.sample(n = 49)
```

```python
print(setosa_sample)
```

```python
new_dataset1 = pd.concat([setosa_sample, iris_virginica], axis=0)
new_dataset2 = pd.concat([setosa_sample, iris_versi_color], axis=0)
```
